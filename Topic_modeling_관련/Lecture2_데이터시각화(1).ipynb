### 볼린저 밴드 그려보자

#### 20일 평균은?

```python
work_df['종가'].rolling(20).mean()
```

#### 20일 표준 편차는?

```python
work_df['종가'].rolling(20).std()
```

```python
work_df['20일평균'] = work_df['종가'].rolling(window=20).mean()
work_df['std'] = work_df['종가'].rolling(window=20).std()
work_df['상단'] = work_df['20일평균'] + 2 * work_df['std']
work_df['하단'] = work_df['20일평균'] - 2 * work_df['std']

work_df
```

```python
plt.figure(figsize=(15,7))
#plt.rcParams['font.family'] = "AppleGothic"

work_df['종가'].plot()
work_df['20일평균'].plot()
work_df['상단'].plot()
work_df['하단'].plot()
plt.show() #추가
```

```python
plt.figure(figsize=(15,7))
work_df['종가'].plot()
work_df['20일평균'].plot(color='red', linewidth='2')
work_df['상단'].plot(color='gray', linestyle=':')
work_df['하단'].plot(color='gray', linestyle=':')
plt.show() #추가
```

```python
plt.figure(figsize=(15,7))
work_df['종가'].plot(color = '#5A7785', label='종가')
# 색상코드 선택은 어떻게 할까?

work_df['20일평균'].plot(linewidth='2', color='#B14824', label='20일 평균')
work_df['상단'].plot(color='gray', linestyle='dashed', label='upper band')
work_df['하단'].plot(color='gray', linestyle='dashed', label='lower band')

plt.legend()
plt.grid()

plt.fill_between(work_df.index, work_df['하단'], work_df['상단'], color='green', alpha=0.05)
# fillbtween은 어떻게 찾았을까?

plt.show() #추가
```

#### 최고가 구하기

```python
from datetime import datetime
from datetime import timedelta

max_price = work_df['종가'].max()   # 최고가(y축)
max_date = work_df[work_df['종가']==max_price].index[-1]  # 최고가를 형성한 날짜(x축)

min_price = work_df['종가'].min()   # 최저가(y축)
min_date = work_df[work_df['종가']==min_price].index[-1]  # 최저가를 형성한 날짜(x축)

```

```python
plt.figure(figsize=(15,7))
work_df['종가'].plot(color = '#465678')
work_df['20일평균'].plot(color='red', linewidth = '2')
work_df['상단'].plot(color = 'gray', linestyle = 'dashed', label = 'upper band')
work_df['하단'].plot(color = 'gray', linestyle = 'dashed', label = 'lower band')

plt.scatter(max_date, max_price, s = 100, c = 'red') # 최고가인 날짜에 점을 찍는 것
plt.text(max_date + timedelta(days=4), max_price, '최고가({0})'.format(max_date), fontsize=14,  # 최고가인 날짜 표시를 하는 것
         verticalalignment='top')


plt.scatter(min_date, min_price, s = 100, c = 'red') # 최저가인 날짜에 점을 찍는 것
plt.text(min_date + timedelta(days=4), min_price, '최저가({0})'.format(min_date), fontsize=14, # 최저가인 날짜 표시를 하는 것
         verticalalignment='top')


plt.legend(fontsize=14)
plt.fill_between(work_df.index, work_df['하단'], work_df['상단'], color = 'green', alpha = 0.05)
plt.show()
```

## 이전 상장 이벤트 반영

#### 임시주주총회 결과(2017-09-29)
* 제1호 의안: 코스닥시장 조건부 상장 폐지 및 유가증권시장 이전 상장 결의의 건 ☞원안대로 가결




```python
event_list = [{'date' :'2017-08-21', 'name' : '주주총회소집결의'}, 
              {'date' :'2017-09-29', 'name' : '유가증권 이전상장 가결'},
              {'date' :'2018-02-07', 'name' : 'KOSDAQ 상장폐지 결정'}]
```

#### Str to Datetime

```python
from dateutil import parser
dt = parser.parse(event_list[0]['date'])
dt

a = datetime.strptime(event_list[0]['date'], "%Y-%M-%d")
a
```

```python
work_df = work_df['2017-08-05':]
```

```python
plt.figure(figsize=(15,7))
work_df['종가'].plot(color = '#465678',  figsize  =(13,10))
work_df['20일평균'].plot(color='red', linewidth = '2')
work_df['상단'].plot(color = 'gray', linestyle = 'dashed', label = 'upper band')
work_df['하단'].plot(color = 'gray', linestyle = 'dashed', label = 'lower band')

plt.scatter(max_date, max_price, s = 100, c = 'red')
plt.text(max_date + timedelta(days=4), max_price, '최고가({0})'.format(max_date), fontsize=14,
         verticalalignment='top')


for event in event_list :
    dt = parser.parse(event['date']) # 날짜 형식으로 변경
    plt.scatter(dt, work_df.loc[dt, '종가'], s = 40, c = 'blue') # 해당 날짜에 점을 찍어주는 것
    plt.text(dt + timedelta(days=4), work_df.loc[dt, '종가'], '{0}({1})'.format(event['name'],event['date']), fontsize=14,
         verticalalignment='top') # 해당 날짜에 공시 이벤트 설명
    plt.axvline(event['date'], color = 'green', linestyle = '--') # 세로축 그어주기

plt.title("셀트리온 이전상장 EVENT", fontsize=16)
    
plt.legend(fontsize=14)
plt.fill_between(work_df.index, work_df['하단'], work_df['상단'], color = 'green', alpha = 0.05)

plt.savefig('graph.png')
```
