# More! Linear Regression!

```
import keras
import numpy

x = numpy.array(range(0,20)) 
y = x * 2 -1 
print(x)
print(y)
```

```
# 혹시 이미 그려둔 그래프가 있다면 날려줘!
keras.backend.clear_session()

# model에 순차적으로 레이어를 쌓아가겠다는 의도!
model = keras.models.Sequential()

# 인풋을 받아, weight을 곱하고, bias를 더해주는거 까지 한 번에 해주는 거
lc = keras.layers.Dense(1,input_shape=(1,))

# model에 위의 블록을 차근 차근 꽂는다.
model.add(lc)

# 컴파일 해주렴
model.compile(loss = 'mse', optimizer = 'sgd')
model.summary()
```

```
# 구경!

from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

SVG(model_to_dot(model, show_shapes = True ).create(prog='dot', format='svg') )
```

```
# 데이터를 넣어서 학습시키자!
model.fit(x[:15], y[:15], epochs=1000, verbose=0)
```

```
# 결과 출력해줘!
print(y[15:])
print(model.predict(x[15:]))
```

## extract weights and bias

역으로 모델 구조를 그려보자

```
weights = model.layers[0].get_weights()[0]
biases = model.layers[0].get_weights()[1]

print(weights, biases)
```

# Logistic Regression More!

```
'''
matplolib inline 명령어를 통해서
matplot으로 그리는 플롯들을 주피터 노트북 내에서 볼 수 있게 해준다.
포맷을 retina로 바꾸면 그래프의 화질이 훨씬 좋아진다.
'''
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

'''
라이브러리들을 불러오자.
'''
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
import keras

import random as rd

from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

### Data Loading & Preparation

[쉬운데이터](https://en.wikipedia.org/wiki/Iris_flower_data_set)로 간다!

![sepal과 petal](https://www.math.umd.edu/~petersd/666/html/iris_with_labels.jpg)

아이리스 데이터!

```
sns.set(style='ticks', color_codes=True)
iris = sns.load_dataset('iris')
plot = sns.pairplot(iris, hue = 'species', palette = 'husl')
```

```
iris.head()
```

```
iris['species'].unique()
```

```
X = iris.iloc[:,0:4].values
y = iris.iloc[:,4].values

encoder =  LabelEncoder()
y1 = encoder.fit_transform(y)
Y = pd.get_dummies(y1).values

print(X[:3])
print(Y[:3])
```

```
train_x, test_x, train_y, test_y = train_test_split(X, Y, 
                                                    test_size=0.2, 
                                                    random_state=1907) 
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

### Sequential을 이용하여 모델링을 할거야! (추천)

```
# 혹시 이미 그려둔 그래프가 있다면 날려줘!
keras.backend.clear_session()

# model에 순차적으로 레이어를 쌓아가겠다는 의도!
model = keras.models.Sequential()

# 인풋을 받아, weight을 곱하고, bias를 더해주고
# softmax까지 씌우는걸 한번에 하자!
sm = keras.layers.Dense(3, input_shape=(4,),
                       activation='softmax')

# model에 위에꺼 lc를 집어 넣을거야. (최초의 레이어)
model.add(sm)

# 컴파일 해주렴
model.compile(loss = 'categorical_crossentropy', optimizer = 'sgd',
              metrics =['accuracy'])

model.summary()
```

```
# 구경!

from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

SVG(model_to_dot(model, show_shapes = True ).create(prog='dot', format='svg') )
```

```
# 데이터를 넣어서 학습시키자!
model.fit(train_x, train_y, epochs=500, verbose=1)
```

```
pred_train = model.predict(train_x)
pred_test = model.predict(test_x)
```

```
pred_train[:3]
```

```
train_y[:3]
```

```
pred_tr_class = np.argmax(pred_train, axis=1)
real_tr_class = np.argmax(train_y, axis=1)

pred_tr_class[:3],  real_tr_class[:3]
```

```
classes = iris['species'].unique().tolist()
```

```
from sklearn.metrics import classification_report, confusion_matrix

print('트레이닝 셋')
print('precision : A라고 예측한 것 중 진짜 A의 비율')
print('recall : 실제 A중, A라고 예측 된 것의 비율')
print(classification_report(real_tr_class, pred_tr_class, target_names = classes ))
print(confusion_matrix(real_tr_class, pred_tr_class).T)
```

```
sns.set(style='ticks', color_codes=True)
iris = sns.load_dataset('iris')
plot = sns.pairplot(iris, hue = 'species', palette = 'husl')
```

```
pred_te_class = np.argmax(pred_test, axis=1)
real_te_class = np.argmax(test_y, axis=1)

pred_te_class[:3],  real_te_class[:3]
```

```
print('트레이닝 셋')
print('precision : A라고 예측한 것 중 진짜 A의 비율')
print('recall : 실제 A중, A라고 예측 된 것의 비율')
print(classification_report(real_te_class, pred_te_class, target_names = classes ))
print(confusion_matrix(real_te_class, pred_te_class).T)
```
